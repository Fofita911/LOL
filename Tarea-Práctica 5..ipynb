```
r = 5
N = 1000

# Inicializar array para media móvil
media_movil = np.zeros(N)

for k in range(r, N - r):
    media_movil[k] = np.mean(manchas[k - r : k + r + 1])

# Para los extremos donde no hay datos suficientes, podemos dejar los datos originales o nan
media_movil[:r] = np.nan
media_movil[N-r:] = np.nan

plt.figure(figsize=(12, 5))
plt.plot(mes[:N], manchas[:N], label='Datos originales')
plt.plot(mes[:N], media_movil, color='red', label='Media móvil (r=5)')
plt.title('Datos originales y media móvil de manchas solares (primeros 1000 datos)')
plt.xlabel('Mes')
plt.ylabel('Número de manchas solares')
plt.grid(True)
plt.legend()
plt.show()
```

# 3. Caminatas aleatorias
a) Escribe un programa que simule una caminata aleatoria para N pasos y haz una gráfica de las posiciones de la caminata vs el tiempo. Asimismo, grafica un histograma de la frecuencia de dichas posiciones. Puedes considerar l = 1, para simplificar las cosas.

```
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
N = 1000  # número de pasos
p = 0.5   # probabilidad de ir a la derecha
q = 1 - p
l = 1     # longitud del paso

# Generar caminata
pasos = np.random.choice([-l, l], size=N, p=[q, p])
posiciones = np.cumsum(pasos)
tiempo = np.arange(1, N + 1)

# Gráfica de la caminata
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(tiempo, posiciones)
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Caminata aleatoria 1D')

# Histograma de posiciones
plt.subplot(1, 2, 2)
plt.hist(posiciones, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Posición')
plt.ylabel('Frecuencia')
plt.title('Histograma de posiciones')
plt.tight_layout()
plt.show()
```

b) Ahora, convierte tu programa del ejercicio anterior en una función y úsala para generar 100 caminatas aleatorias y haz una sola gráfica de las posiciones de todas las caminatas vs el tiempo t. Asimismo, grafica un único histograma de la frecuencia de las posiciones, de todas las caminatas. ¿Qué es lo que observas?, ¿cuál es la diferencia con  lo que obtuviste en el ejercicio anterior?

```
import numpy as np
import matplotlib.pyplot as plt

def caminata_aleatoria(N, p=0.5, l=1):
    pasos = np.random.choice([-l, l], size=N, p=[1-p, p])
    return np.cumsum(pasos)

# Parámetros
N = 1000
num_caminatas = 100
t = np.arange(1, N + 1)
todas_posiciones = []

# Graficar todas las caminatas
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
for _ in range(num_caminatas):
    pos = caminata_aleatoria(N)
    todas_posiciones.append(pos)
    plt.plot(t, pos, alpha=0.3)
plt.title("100 Caminatas aleatorias 1D")
plt.xlabel("Tiempo")
plt.ylabel("Posición")

# Histograma total de posiciones finales
plt.subplot(1, 2, 2)
todas_posiciones_np = np.array(todas_posiciones)
pos_finales = todas_posiciones_np[:, -1]
plt.hist(pos_finales, bins=30, color='orange', edgecolor='black')
plt.title("Histograma de posiciones finales")
plt.xlabel("Posición final")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```

c) Finalmente, generaliza tu programa para crear una función para simular una (o varias) caminatas aleatorias en 2D y 3D, y úsalas para graficar las posiciones (x, y) o (x, y, z) de una o varias caminatas.

```
from mpl_toolkits.mplot3d import Axes3D  # para la caminata 3D

def caminata_aleatoria_2D(N):
    pasos = np.random.choice(['arriba', 'abajo', 'izq', 'der'], size=N)
    x, y = 0, 0
    posiciones = [(x, y)]
    for paso in pasos:
        if paso == 'arriba':
            y += 1
        elif paso == 'abajo':
            y -= 1
        elif paso == 'izq':
            x -= 1
        elif paso == 'der':
            x += 1
        posiciones.append((x, y))
    return np.array(posiciones)

def caminata_aleatoria_3D(N):
    pasos = np.random.choice(['x+', 'x-', 'y+', 'y-', 'z+', 'z-'], size=N)
    x, y, z = 0, 0, 0
    posiciones = [(x, y, z)]
    for paso in pasos:
        if paso == 'x+': x += 1
        elif paso == 'x-': x -= 1
        elif paso == 'y+': y += 1
        elif paso == 'y-': y -= 1
        elif paso == 'z+': z += 1
        elif paso == 'z-': z -= 1
        posiciones.append((x, y, z))
    return np.array(posiciones)

# Caminata 2D
pos_2D = caminata_aleatoria_2D(500)
plt.figure(figsize=(6, 6))
plt.plot(pos_2D[:, 0], pos_2D[:, 1])
plt.title("Caminata Aleatoria en 2D")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()

# Caminata 3D
pos_3D = caminata_aleatoria_3D(500)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(pos_3D[:, 0], pos_3D[:, 1], pos_3D[:, 2])
ax.set_title("Caminata Aleatoria en 3D")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

# 4. La gráfica de Feigenbaum (caos determinista)
a) Apoyate en el programa que vimos en clase y escribe un programa que muestre
el comportamiento del mapeo logístico mediante una gráfica.

```
import matplotlib.pyplot as plt

# Rango de valores para r
r_values = []
x_values = []

# Itera valores de r de 1.0 a 4.0 con paso de 0.01
for r in [i * 0.01 for i in range(100, 401)]:
    x = 0.5  # Valor inicial de x

    # Primeras 1000 iteraciones para estabilizar
    for _ in range(1000):
        x = r * x * (1 - x)

    # Otras 1000 iteraciones para graficar
    for _ in range(1000):
        x = r * x * (1 - x)
        r_values.append(r)
        x_values.append(x)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(r_values, x_values, 'k.', markersize=0.5)
plt.xlabel('r')
plt.ylabel('$x_\\infty$')
plt.title('Gráfica de Feigenbaum (mapa logístico)')
plt.grid(True)
plt.show()
```

# 5. El conjunto de Mandelbrot
a) Escribe un programa para crear una imagen del conjunto de Mandelbrot realizando la iteración para todos los valores de  
**c = x + iy** en una cuadrícula de **N × N** que abarque la región donde **−2 ≤ x ≤ 2** y **−2 ≤ y ≤ 2**.  Haz una **gráfica de densidad** (density plot) en la que los puntos de la cuadrícula **dentro del conjunto de Mandelbrot estén coloreados en negro** y los de afuera **en blanco**.

```
import numpy as np
import matplotlib.pyplot as plt

# Tamaño de la cuadrícula
N = 500
# Rango del plano complejo
x = np.linspace(-2, 2, N)
y = np.linspace(-2, 2, N)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y

# Inicializar Z y matriz del resultado
Z = np.zeros_like(C)
mandelbrot = np.ones(C.shape, dtype=bool)

# Número de iteraciones
iteraciones = 100

for i in range(iteraciones):
    Z = Z**2 + C
    mandelbrot &= (np.abs(Z) <= 2)

# Graficar el resultado
plt.figure(figsize=(8, 8))
plt.imshow(mandelbrot, cmap='binary', extent=(-2, 2, -2, 2))
plt.title('Conjunto de Mandelbrot (Blanco y Negro)')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.grid(False)
plt.show()
```

En lugar de colorear los puntos solo en blanco o negro, **colorea los puntos de acuerdo con el número de iteraciones de la ecuación** antes de que **|zₙ| sea mayor que 2**  (o bien el número máximo de iteraciones si es que |zₙ| nunca llega a ser mayor que 2).  Si usas alguno de los esquemas más coloridos que Python proporciona para las gráficas de densidad, como `"hot"` o `"jet"`, puedes crear algunas imágenes muy espectaculares.  Otra variante interesante es **colorear según el logaritmo del número de iteraciones**,lo que ayuda a revelar parte de la estructura más fina fuera del conjunto.

```
import numpy as np
import matplotlib.pyplot as plt

# Resolución
N = 500

# Crear el plano complejo
x = np.linspace(-2, 2, N)
y = np.linspace(-2, 2, N)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y

# Inicializar Z y matriz de iteraciones
Z = np.zeros_like(C)
iteraciones_max = 100
diverge_map = np.zeros(C.shape)

for i in range(iteraciones_max):
    Z = Z**2 + C
    mask = (np.abs(Z) <= 2)
    diverge_map += mask.astype(int)

# Graficar con un mapa de colores
plt.figure(figsize=(8, 8))
plt.imshow(diverge_map, extent=(-2, 2, -2, 2), cmap='hot')
plt.title('Conjunto de Mandelbrot (Coloreado por iteraciones)')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.grid(False)
plt.colorbar(label='Número de iteraciones')
plt.show()
```
